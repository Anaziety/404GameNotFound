```python
# Using 1 for PlayerX, 2 for PlayerY, 0 for empty spaces
PlayerX, PlayerY = 1, 2

# To check if there are moves left
def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return True
    return False

# Evaluating score to find if its true or not
def CheckScore(b):
    # Row matching
    for row in range(3):
        if b[row][0] == b[row][1] and b[row][1] == b[row][2]:
            if b[row][0] == PlayerX:
                return 10
            elif b[row][0] == PlayerY:
                return -10

    # Column matching
    for col in range(3):
        if b[0][col] == b[1][col] and b[1][col] == b[2][col]:
            if b[0][col] == PlayerX:
                return 10
            elif b[0][col] == PlayerY:
                return -10

    # Diagonal checking
    if b[0][0] == b[1][1] and b[1][1] == b[2][2]:
        if b[0][0] == PlayerX:
            return 10
        elif b[0][0] == PlayerY:
            return -10

    if b[0][2] == b[1][1] and b[1][1] == b[2][0]:
        if b[0][2] == PlayerX:
            return 10
        elif b[0][2] == PlayerY:
            return -10

    return 0

def minimax(board, depth, isMax):
    score = CheckScore(board)

    # PlayerX wins here
    if score == 10:
        return score
    # PlayerY wins here
    if score == -10:
        return score
    # Game ended or draw
    if not isMovesLeft(board):
        return 0

    # If Maximize
    if isMax:
        best = -1000

        for i in range(3):
            for j in range(3):
                if board[i][j] == 0:
                    board[i][j] = PlayerX
                    best = max(best, minimax(board, depth + 1, not isMax))
                    board[i][j] = 0
        return best

    # Else part to Minimize
    else:
        best = 1000

        for i in range(3):
            for j in range(3):
                if board[i][j] == 0:
                    board[i][j] = PlayerY
                    best = min(best, minimax(board, depth + 1, not isMax))
                    board[i][j] = 0
        return best

def findnextMove(board):
    nextVal = -1000
    nextMove = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                board[i][j] = PlayerX
                moveVal = minimax(board, 0, False)
                board[i][j] = 0

                if moveVal > nextVal:
                    nextMove = (i, j)
                    nextVal = moveVal

    print("The value of the best Move is:", nextVal)
    print()
    return nextMove

# Main
playerX_game = [
    [1, 0, 0],
    [0, 0, 0],
    [0, 2, 0]
]

nextMove = findnextMove(playerX_game)
print("THE BEST MOVE:", nextMove[0], nextMove[1])
```
